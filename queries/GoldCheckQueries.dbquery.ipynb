```sql
select * from `workspace`.`default`.`fact_inspection_violation` as fiv
left join `workspace`.`default`.`dim_violation` using (violation_key)
-- where (fiv.violation_points is not null or fiv.violation_slot is not null)
where fiv.source_city = 'Chicago'
 limit 100;

-- Count of risk levels in dim_estableshement
select count(*), risk_level from workspace.default.dim_establishment
group by risk_level;

-- Check what risk level 'all'
select * from workspace.default.dim_establishment
where risk_level = 'All';

-- check long/lang values in location
select count(*) from workspace.default.dim_location
where longitude is not null and latitude is not null;

-- get long/land values range 
-- Longitude values should be between -180 and 180.
-- Latitude values should be between -90 and 90
select min(longitude), max(longitude), min(latitude), max(latitude) from workspace.default.dim_location
where longitude is not null and latitude is not null;

-- Find bad rows
-- Check for coordinates outside Chicago and Dallas reasonable ranges
-- Chicago: lat ~41.6-42.1, long ~-87.9 to -87.5
-- Dallas: lat ~32.6-33.1, long ~-97.0 to -96.4

SELECT city, state, latitude, longitude, COUNT(*) as cnt
FROM workspace.default.dim_location
WHERE longitude IS NOT NULL 
  AND latitude IS NOT NULL
  AND NOT (
    (latitude BETWEEN 41.5 AND 42.2 AND longitude BETWEEN -88.0 AND -87.4)  -- Chicago
    OR
    (latitude BETWEEN 32.5 AND 33.2 AND longitude BETWEEN -97.1 AND -96.3)  -- Dallas
  )
GROUP BY city, state, latitude, longitude
ORDER BY cnt DESC
LIMIT 20;
```
