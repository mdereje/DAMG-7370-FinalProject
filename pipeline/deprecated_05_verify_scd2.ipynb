```python
from pyspark.sql.functions import col

scd2 = spark.table("workspace.default.dim_violation_scd2")

print(f"dim_violation_scd2: {scd2.count()} rows")
print(f"Current records: {scd2.where(col('__END_AT').isNull()).count()}")
print(f"Historical records: {scd2.where(col('__END_AT').isNotNull()).count()}")

print("\n=== SCHEMA ===")
scd2.printSchema()

print("\n=== SAMPLE RECORDS ===")
display(scd2.orderBy("source_city", "violation_code").limit(20))
```
