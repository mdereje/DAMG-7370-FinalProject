### Data Exploration - Food Inspection Chicago

```python
# load the chicago file
chicago_df = spark \
    .read.format("csv") \
    .option("inferSchema","true") \
    .option("header","true") \
    .option("sep",",") \
    .load("/Volumes/workspace/final_project_raw/food_inspection/Food_Inspections_Chicago.csv")
```

```python
chicago_df.printSchema()
```

```python
chicago_df.createOrReplaceTempView("food_inspection_chicago")
```

```python
spark.sql("select * from food_inspection_chicago limit 10").display()
```

```python
spark.sql("select count(`Inspection ID`) from food_inspection_chicago").display()
```

```python
spark.sql("select count(distinct `Inspection ID`) from food_inspection_chicago").display()
```

```python
spark.sql("select `DBA Name`, `License #`, `Address`, `Facility Type`, `Zip` from food_inspection_chicago where `DBA Name` like '%MCD%' order by `Address` desc limit 100").display()
```

```python
spark.sql("select distinct Risk from food_inspection_chicago").display()
```

```python
spark.sql("select * from food_inspection_chicago where Risk = 'All'").display()
```

```python
spark.sql("select * from food_inspection_chicago where Results = 'Pass' limit 20").display()
```

```python
spark.sql("select distinct `Facility Type` from food_inspection_chicago").display()
```

```python
spark.sql("select * from final_project_silver.chicago_quarantine limit 10").display()
```

```python
spark.sql("select * from food_inspection_chicago where latitude = '2 DOORS NOT CLOSING SAME AREA AND EXIT DOOR(S) (N)'").display()
```

### Data Exploration - Food Inspection Dallas

```python
# load the dallas file
dallas_df = spark \
    .read.format("csv") \
    .option("inferSchema","true") \
    .option("header","true") \
    .option("sep",",") \
    .load("/Volumes/workspace/final_project_raw/food_inspection/Food_Inspections_Dallas.csv")
```

```python
dallas_df.printSchema()
```

```python
dallas_df.createOrReplaceTempView("food_inspection_dallas")
```

```python
spark.sql("select * from food_inspection_dallas limit 10").display()
```
